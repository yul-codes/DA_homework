### Задание 1

Соберите всю информацию о студентах в одну таблицу df. В получившейся таблице должна быть информация и оценки всех студентов из всех групп. Напечатайте несколько строк таблицы для демонстрации результата

```python
# ваш код здесь

st_info = pd.DataFrame()
grades = pd.DataFrame()

for i in range(10):
    st_info = pd.concat(
        [st_info, pd.read_csv(f"students/Students_info_{i}.csv")], ignore_index=True
    )
    grades = pd.concat(
        [grades, pd.read_csv(f"students/Students_marks_{i}.csv")], ignore_index=True
    )
```

```python
# смотрим структуру датафреймов, чтобы определить, есть ли общие колонки для слияния

st_info.head()
```

### Задание 14

Сколько студентов успешно сдали экзамен по математике?

Создайте новый столбец в таблице `df` под названием `Math_PassStatus` и запишите в него F, если студент не сдал экзамен по математике (балл за экзамен < passmark), и P иначе.

Посчитайте количество студентов, сдавших и не сдавших экзамен по математике.

Сделайте аналогичные шаги для экзаменов по чтению и письму.

```python
# ваш код здесь
# через функцию для минимизации копипасты
def results_breakdown(df, col, passmark=50):
    col_upd = col.replace(" score", "").capitalize()
    status_col = col_upd + "_PassStatus"
    df[status_col] = ["F" if row < passmark else "P" for row in df[col]]
    passed = df[status_col].value_counts()[
        "P"
    ]  # или df[df[status_col] == 'P'].shape[0]
    failed = df[status_col].value_counts()[
        "F"
    ]  # или df[df[status_col] == 'F'].shape[0]
    return f"Passed {col_upd}: {passed} / Failed {col_upd}: {failed}"
```

```python
print(results_breakdown(df, "math score"))
print(results_breakdown(df, "reading score"))
print(results_breakdown(df, "writing score"))
```

### Задание 16

Переведем баллы в оценки

**Система перевода баллов в оценки**
*   [90;100] = A
*   [80;90) = B
*   [70;80) = C
*   [60;70) = D
*   [50;60) = E
*   [0;50) = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец `Grade` и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

```python
# для наглядности самовольно добавляю стоблец со средней оценкой, по моему мнению (человека, который с подобным работал 4 года), он вполне уместен и полезен
df["Avg_Score"] = (
    df[["math score", "reading score", "writing score"]].mean(axis=1).round(2)
)

df
```

```python
def GetGrade(average_mark):
    # ваш код здесь
    # создание словаря с диапазонами оценок
    grading_scale = {
        "A": (90, 100),
        "B": (80, 89),
        "C": (70, 79),
        "D": (60, 69),
        "E": (50, 59),
        "F": (0, 49),
    }
    for grade, (band_bottom, band_top) in grading_scale.items():
        if band_bottom <= float(average_mark) <= band_top:
            return grade
    return "F"
    # я не очень поняла, почему присваивание 'F' не происходит на уровне условия несмотря на то, что этот диапазон есть в словаре, но таким образом хотя бы корректно сработало в итоге


# ваш код здесь
df["Grade"] = df["Avg_Score"].apply(GetGrade)

df
```
